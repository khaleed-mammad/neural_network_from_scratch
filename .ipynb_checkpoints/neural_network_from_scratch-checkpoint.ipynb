# Building Value Class

```python
# This is a graph visualization code from micrograd (https://github.com/karpathy/micrograd/blob/master/trace_graph.ipynb)

from graphviz import Digraph

def trace(root):
    nodes, edges = set(), set()
    def build(v):
        if v not in nodes:
            nodes.add(v)
            for child in v._prev:
                edges.add((child, v))
                build(child)
    build(root)
    return nodes, edges

def draw_dot(root, format='svg', rankdir='LR'):
    """
    format: png | svg | ...
    rankdir: TB (top to bottom graph) | LR (left to right)
    """
    assert rankdir in ['LR', 'TB']
    nodes, edges = trace(root)
    dot = Digraph(format=format, graph_attr={'rankdir': rankdir}) #, node_attr={'rankdir': 'TB'})

    for n in nodes:
        dot.node(name=str(id(n)), label = "{ %s | data %.2f | grad %.3f }" % (n.label, n.data, n.grad), shape='record')
        if n._op:
            dot.node(name=str(id(n)) + n._op, label=n._op)
            dot.edge(str(id(n)) + n._op, str(id(n)))

    for n1, n2 in edges:
        dot.edge(str(id(n1)), str(id(n2)) + n2._op)

    return dot
```

```python
class Value:
    
    def __init__(self, data, _children=(), _op='', label = ''):
        self.data = data
        self._prev = set(_children) # It is used when we do operations (add, mul), we want to know values we added (multiplied)
        self._op = _op # It shows how the value is produced (by adding or multiplying)
        self.label = label
        self.grad = 0.0
        self._backward = lambda: None
        
    def __repr__(self):
        return f"Value({self.data})"
    
    
    # Since pythin does not know how to add two objects, we need to create __add__ magic function.
    # In this case, when we call a+b, it will be interpreted as a.__add__(b) where a is self, b is other
    def __add__(self,other):
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data + other.data, (self,other), _op = '+')
        
        def _backward():
            self.grad += out.grad * 1.0
            other.grad += out.grad * 1.0
            
        out._backward = _backward
        return out
    
    
    #Same for multiplication
    def __mul__(self, other):
        
        other = other if isinstance(other, Value) else Value(other)
        out = Value(self.data * other.data, (self,other), _op = '*')
        
        def _backward():
            self.grad += out.grad * other.data
            other.grad += out.grad * self.data
        out._backward = _backward
        return out
    
    
    def __rmul__(self,other): #It just does __mul__(other,self) instead of __mul__(self, other)
        return self*other
    
    def __pow__(self, other):
        assert isinstance(other, (int, float)), "only supporting int/float powers for now"
        out = Value(self.data**other, (self,), f'**{other}')
        
        def _backward():
            self.grad += other * (self.data ** (other - 1)) * out.grad
        out._backward = _backward
        return out
    
    def __truediv__(self, other): # self / other
        return self * other**-1

    def __neg__(self): # -self
        return self * -1

    def __sub__(self, other): # self - other
        return self + (-other)

    def __radd__(self, other): # other + self
        return self + other
    
    
    
    
    # We need activation function to introduce non-linearity to our network. Without activation function,
    # a neural network would simply behave like a linear regression model, regardless of how many layers it has.
    # We use Tanh (Hyperbolic Tangent) just to demonstrat
    
    def tanh(self):
        x = self.data
        t = (math.exp(2*x) - 1) / (math.exp(2*x) + 1)
        out = Value(t, (self, ), label = 'tanh')
        
        def _backward():
            self.grad += (1 - math.pow(t,2)) * out.grad
            
        out._backward = _backward
        return out
        
    def backward(self):
        topo =[]
        visited = set()
        def build_topo(v):
            if v not in visited:
                visited.add(v)
                for child in v._prev:
                    build_topo(child)
                topo.append (v)
        
        
        build_topo(self)
        self.grad = 1.0
        for node in reversed(topo):
            node._backward()
```

```python
a = Value(2.0, label = 'a')
b = Value(-3.0, label = 'b')
c = Value(10.0, label = 'c')
e = a*b; e.label = 'e'
d = e+c; d.label = 'd'
f = Value(-2.0,label='f')
L = d * f; L.label = 'Loss'
```

```python
draw_dot(L)
```

# Manual Backpropogation

```python
L.grad = 1.0 # It is because derivative of L with respect to L is 1

d.grad = f.data # L = d*f, so der. of L with respect to d is f
f.grad = d.data

# Now, comes crucial part. We need to find c.grad (also e.grad)
# c.grad means derivative of L with respect to c. 
# Here, we need to use chain rule. Because, L is not directly depend on c.
# However, L depends on d which directly depends on c.
# Therefore, dL/dc (which is c.grad) = dL/dd * dd/dc (according to chain rule)
# Also, dd/dc is 1 as d = c + e

c.grad = d.grad * 1
e.grad = d.grad * 1

# Same rules for a and b (dL/da = dL/de * de/da)
# de/da is b

a.grad = e.grad * b.data
b.grad = e.grad * a.data
```

```python
draw_dot(L)
```

# Single optimization step

```python
# When we say optimization, our purpose is to bring Loss near to 0. For this, we update our values.
# When we update our values, we update them in the direction of gradient by small value (learning rate) * gradient
# We update a, b, c, and f. e and d will automotically we updated as they depend on other values.

a.data += 0.01 * a.grad
b.data += 0.01 * b.grad
c.data += 0.01 * c.grad
f.data += 0.01 * f.grad

# Now, we need to do forward pass which means calculate e, d and L based on new values

e = a * b; e.label = 'e'
d = c + e; d.label = 'd'
L = d * f; L.label = 'Loss'

# We expect our Loss to come to near to 0, meaning more than -8.
```

```python
draw_dot(L)
```

# Simple Neural Network

```python
# Inputs

x1 = Value(2.0, label = 'x1')
x2 = Value(0.0, label = 'x2')

#Weights

w1 = Value(-3.0,label = 'w1')
w2 = Value(1.0, label = 'w2')

# Bias of the neuron
b = Value(6.8813735870195432, label = 'b')

# Forward Pass (with activation function tanh)

x1w1 = x1*w1; x1w1.label = 'x1w1'
x2w2 = x2*w2; x2w2.label = 'x2w2'
x1w1x2w2 = x1w1 + x2w2; x1w1x2w2.label = 'x1w1 + x2w2'
n = x1w1x2w2 + b; n.label = 'n'
o = n.tanh(); o.label = 'tanh'
```

```python
draw_dot(o)
```

```python
#MANUAL BACKPROPAGATION

# Derivative of o with respect to o is 1.
o.grad = 1

# Derivative of o with respect to n:
# o = tanh(n). derivative of tanh(n) = 1 - (tanh(x))^2

n.grad = 1 - math.pow(o.data,2)

#Others are just a chain rule

x1w1x2w2.grad = n.grad * 1
b.grad = n.grad * 1

x1w1.grad = x1w1x2w2.grad * 1
x2w2.grad = x1w1x2w2.grad * 1

x2.grad = x2w2.grad * w2.data
w2.grad = x2w2.grad * x2.data


x1.grad = x1w1.grad * w1.data
w1.grad = x1w1.grad * x1.data
```

```python
draw_dot(o)
```

# Neural Net Library

```python
class Neuron:
    
    def __init__(self,nin): # nin is num. of inputs for our network
        self.nin = nin
        self.weights = [Value(random.uniform(-1,1)) for _ in range(nin)]
        self.bias = Value(random.uniform(-1,1))
        
    def __call__(self,x): #__call__ allows an instance of a class to be called like a function (we will do forward pass here)
        act = sum((wi*xi for wi,xi in zip(self.weights,x)), self.bias) #It will add b + wixi for all w and x
        out = act.tanh() #non-linearity
        return out
    
    def parameters(self):
        return self.weights + [self.bias]
        
class Layer:
    def __init__(self,nin,nout): # nin is num. of inputs for our network, nout is num. of outputs for our network
        self.neurons = [Neuron(nin) for _ in range(nout)]
        
        
    def __call__(self,x):
        outs = [n(x) for n in self.neurons]
        return outs[0] if len(outs)==1 else outs
    
    def parameters(self):
        params = []
        for neuron in self.neurons:
            ps = neuron.parameters()
            params.extend(ps)
        return params
        
class MLP:
    def __init__(self,nin,nouts): # here, nouts is list of numbers show num. of outputs in each layer (except input)
        sz = [nin] + nouts # It is num. of neurons in each layer (including input)
        self.layers = [Layer(sz[i],sz[i+1]) for i in range(len(nouts))]
    
    def __call__(self,x):
        for layer in self.layers:
            x = layer(x)
        return x
    
    def parameters(self):
        params = []
        for layer in self.layers:
            ps = layer.parameters()
            params.extend(ps)
        return params
```

# Working on artificial and small datasets (+loss function)

```python
# inputs
xs = [
    [2, 3, -1],
    [3, -1, 0.5],
    [0.5, 1, 1],
    [1, 1, -1]
]

#actual outputs
ys = [1, -1, -1, 1]
```

```python
predictions = [n(x) for x in xs]
predictions
```

```python
# Mean Squared Error

loss = sum([(pred - y)**2 for pred, y in zip(predictions, ys)])
loss
```

```python
# We calculate gradients
loss.backward()

# We need to update our values

for p in n.parameters():
    p.data = p.data - 0.01 * p.grad
```

```python
predictions = [n(x) for x in xs]
loss = sum([(pred - y)**2 for pred, y in zip(predictions, ys)])
loss
```

# Training

```python
# inputs
xs = [
    [2, 3, -1],
    [3, -1, 0.5],
    [0.5, 1, 1],
    [1, 1, -1]
]

#actual outputs
ys = [1, -1, -1, 1]

# hyperparameters

ephocs = 10
learning_rate = 0.01

for ephoc in range(ephocs):
    
    #forward pass
    predictions = [n(x) for x in xs]
    loss = sum([(pred - y)**2 for pred, y in zip(predictions, ys)])
    
    #backward pass
    loss.backward()
    
    #optimization
    for p in n.parameters():
        p.data = p.data - learning_rate * p.grad
        p.grad = 0
        
    print(f"Loss after iteration {ephoc+1}: {loss}")
```
